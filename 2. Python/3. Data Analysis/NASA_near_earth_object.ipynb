# NEO - Near Earth Objects by NASA

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

neo_data = pd.read_csv('neo.csv')
numberOfRows, numberOfCol = neo_data.shape

print(f'Total NEO: {numberOfRows:9}')
print(f'Displaying Statistics: {numberOfCol:}')

neo_data.head(5)
```

## Overall Statistics

## Propotion to Sizes among NEOs

```python
def print_doughnut(mlist, mlabel, mtitle):

    fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))
    wedges, texts = ax.pie(mlist, wedgeprops=dict(width=0.5), startangle=-40)

    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    kw = dict(arrowprops=dict(arrowstyle="-"),
            bbox=bbox_props, zorder=0, va="center")

    # label magnify
    x_mag = 0
    y_mag = 0
    for i, p in enumerate(wedges):
        ang = (p.theta2 - p.theta1)/2. + p.theta1
        y = np.sin(np.deg2rad(ang))
        x = np.cos(np.deg2rad(ang))
        horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
        connectionstyle = "angle,angleA=0,angleB={}".format(ang)
        kw["arrowprops"].update({"connectionstyle": connectionstyle})

        x_mag += 1
        y_mag += 0.6
        ax.annotate(mlabel[i], xy=(x, y), xytext=(x_mag*np.sign(x), y_mag*y), horizontalalignment = horizontalalignment, **kw)   # default 1.35 np , 1.4 y

    ax.set_title(mtitle)
    
    plt.show()

print("\n<------------------------------------------------------------------------------------------>\n")
h_6 = hazardous_neo_data[(hazardous_neo_data.est_diameter_max >= 6)]
h_5_6 = hazardous_neo_data[(hazardous_neo_data.est_diameter_max < 6) & (hazardous_neo_data.est_diameter_max >= 5)]
h_4_5 = hazardous_neo_data[(hazardous_neo_data.est_diameter_max < 5) & (hazardous_neo_data.est_diameter_max >= 4)]
h_3_4 = hazardous_neo_data[(hazardous_neo_data.est_diameter_max < 4) & (hazardous_neo_data.est_diameter_max >= 3)]
h_2_1 = hazardous_neo_data[(hazardous_neo_data.est_diameter_max < 2) & (hazardous_neo_data.est_diameter_max >= 1)]
h_1 = hazardous_neo_data[(hazardous_neo_data.est_diameter_max < 1)]

h_title ='Hazardous NEO Sizes in Doughnut Chart (Unit of Meas.: Kilometers)'
hazardous_neo_size_list = [len(h_6), len(h_5_6), len(h_4_5), len(h_3_4), len(h_2_1), len(h_1)]
h_neo_labels = ['Above 6 KM', '5~6 KM', '4~5 KM', '3~4 KM', '1~2 KM', 'Under 1 KM']

hazardous_neo_size_list.reverse()
h_neo_labels.reverse()

h_neo_dict = {
                'Diameter': h_neo_labels,
                'Numbers': hazardous_neo_size_list
             }
h_neo_data = pd.DataFrame(h_neo_dict)

print_doughnut(hazardous_neo_size_list, h_neo_labels, h_title)
print(h_neo_data)


print("\n<------------------------------------------------------------------------------------------>\n")
nh_6 = non_hazardous_neo_data[(non_hazardous_neo_data.est_diameter_max >= 6)]
nh_5_6 = non_hazardous_neo_data[(non_hazardous_neo_data.est_diameter_max < 6) & (non_hazardous_neo_data.est_diameter_max >= 5)]
nh_4_5 = non_hazardous_neo_data[(non_hazardous_neo_data.est_diameter_max < 5) & (non_hazardous_neo_data.est_diameter_max >= 4)]
nh_3_4 = non_hazardous_neo_data[(non_hazardous_neo_data.est_diameter_max < 4) & (non_hazardous_neo_data.est_diameter_max >= 3)]
nh_2_1 = non_hazardous_neo_data[(non_hazardous_neo_data.est_diameter_max < 2) & (non_hazardous_neo_data.est_diameter_max >= 1)]
nh_1 = non_hazardous_neo_data[(non_hazardous_neo_data.est_diameter_max < 1)]


nh_title ='Non-Hazardous NEO Sizes in Doughnut Chart (Unit of Meas.: Kilometers)'
non_hazardous_neo_size_list = [len(nh_6), len(nh_5_6), len(nh_4_5), len(nh_3_4), len(nh_2_1), len(nh_1)]
nh_neo_labels = ['Above 6 KM', '5~6 KM', '4~5 KM', '3~4 KM', '1~2 KM', 'Under 1 KM']

non_hazardous_neo_size_list.reverse()
nh_neo_labels.reverse()

nh_neo_dict = {
                'Diameter': nh_neo_labels,
                'Numbers': non_hazardous_neo_size_list
              }
nh_neo_data = pd.DataFrame(nh_neo_dict)

print_doughnut(non_hazardous_neo_size_list, nh_neo_labels, nh_title)
print(nh_neo_data)

print("\n<------------------------------------------------------------------------------------------>\n")

```
